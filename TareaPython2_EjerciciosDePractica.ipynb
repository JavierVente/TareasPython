**Numeros primos**
- Pide al usuario que ingrese un número. 
- El programa deberá determinar si el número es o no primo e imprimir un mensaje adecuado.

```python
usarSistema = True
while usarSistema:
  print("Por favor ingrese un numero entero positivo mayor o igual a 2 (Presione Enter para salir)")

  numeroIngresado = input()

  #Verificar si el usuario desea salir del sistema
  if numeroIngresado == "":
    usarSistema = False
    break

  #El siguiente bloque se encargara de asegurarse de que el valor ingresado sea un numero entero
  try:
    numeroIngresado = int(numeroIngresado)
  except Exception as e:
    print("El valor ingresado no corresponde a un numero entero positivo mayor o igual a 2")
    print("")
    print("-----------------------------------------")
    continue

  #Aqui se valida que el numero entero sea mayor o igual a 2
  if not numeroIngresado >= 2:
      print("El valor ingresado no corresponde a un numero entero positivo mayor o igual a 2")
      print("")
      print("-----------------------------------------")
      continue
  
  #En esta seccion se intenta dividir al numero ingresado con numeros menores para ver si es primo o no
  #En un principio se asume que es primo, y si se halla una division entera se sale del bucle
  esPrimo = True
  for posibleDivisor in range(2, ((numeroIngresado // 2) + 1)):
    
    division = (numeroIngresado / posibleDivisor)

    if int(division) == division:
      division = int(division)
      esPrimo = False
      print("")
      print(f"{posibleDivisor} divide a {numeroIngresado}: {numeroIngresado} / {posibleDivisor} = {division}")
      break
  
  #La palabra 'no ' se agrega en el caso de que no sea un primo, haciendo negativa a la respuesta final
  negacion = ""
  if not esPrimo:
    negacion = "no "
  
  print("")
  print(f"El numero {numeroIngresado} {negacion}es primo")
  print("")
  print("-----------------------------------------")

```
