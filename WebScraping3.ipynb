## Descargar letras de canciones

Utilizando beautiful soup descargar todas las canciones de [Drexler](https://www.letras.com/jorge-drexler/) que hay en [letras.com](https://www.letras.com/jorge-drexler/)

```python
from bs4 import BeautifulSoup
import requests
import os
```

```python


def descargar_letra_canciones(nombre_artista):
  nombres = nombre_artista.lower().split(" ")
  path_artista = "-".join(nombres)
  url = f"https://www.letras.com/{path_artista}/"

  r = requests.get(url)
  content = r.content
  soup = BeautifulSoup(content)
  
  # Juntar todas las canciones del artista, 
  # y ademas la url donde se encuentran las letras de cada cancion en si
  url_canciones = []
  nombre_canciones = []
  letra_canciones = []
  for cancion in soup.find_all('a', attrs='song-name'):
    nombre_canciones.append(cancion.text)
    url_canciones.append(f'https://www.letras.com/{(cancion.get("href"))}')

  # Buscar la letra de cada cancion
  for url in url_canciones:
    r = requests.get(url)
    content = r.content
    soup = BeautifulSoup(content)

    letra_cancion = []
    letra_container = soup.find('div', attrs='cnt-letra p402_premium')
    for estrofa in letra_container.find_all('p'):
      letra_cancion.append(estrofa.get_text(strip=True, separator='\n').splitlines())
    
    letra_canciones.append(letra_cancion)


  return (imprimir_canciones(nombre_artista, nombre_canciones, letra_canciones))
```

```python
def imprimir_canciones(artista, nombre_canciones, letra_canciones):
  print(f'LETRA DE TODAS LAS CANCIONES ENCONTRADAS DE {artista.upper()}')
  print("")

  if len(nombre_canciones) == len(letra_canciones):
    contador_cancion = 0
    for nombre_cancion in nombre_canciones:
      print(nombre_cancion.upper())
      print("")

      for estrofa in letra_canciones[contador_cancion]:
        for verso in estrofa:
          print(f'  {verso}')

        print("")

      print("")
      print("")
      
      contador_cancion +=1
```

```python
descargar_letra_canciones("hollow coves")
```
